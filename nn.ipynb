```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
```

```python
x = torch.tensor([[-0.7411, -0.5078,-0.3206], 
                       [0.0983, -0.0308, -0.3728],
                       [0.0414,0.2323,-0.2365],
                       [-0.7342,0.4264,2.0237]],
                     requires_grad=True)


print(x)
```

```python
y = torch.FloatTensor([[0.00,1.00,0.00],[1.00,0.00,0.00],
                 [0.00,1.00,0.00],[0.00,0.00,1.00]],)
print(y)
```

```python
#### Creatig Module
class NeuralNetwork(nn.Module):
    def __init__(self):
        w1 = [[1.6035, -1.5062,0.2761], [1.2347, -0.4446, -0.2612], 
              [-0.2296,-0.1559,-0.4434]]
        b1 = [0.3919,-1.2507,-0.9480]
        w2 = [[0.0125,1.2424,0.3503],[-3.0292,-1.0667,-0.0290],
              [-0.457,0.9337,0.1825]]
        b2 = [-1.5651,-0.0845,1.6039]
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(3,3)
        self.fc2 = nn.Linear(3,3)
        self.fc1.bias = nn.Parameter(torch.FloatTensor(b1))
        self.fc1.weight = nn.Parameter(torch.FloatTensor(w1))
        self.fc2.weight = nn.Parameter(torch.FloatTensor(w2))
        self.fc2.bias = nn.Parameter(torch.FloatTensor(b2))
        
    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
####Instantiate the NeuralNetwork
model = NeuralNetwork()
print(model.fc1.weight)
print(model.fc2.weight)
print("-------------")
print(model.fc1.bias)
print(model.fc2.bias)
```

```python
model.parameters
```

```python
####Backward Propagationï¼Œ Define the loss function and optimizer
loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
```

```python
epochs = 20
final_loss = []
for i in range(epochs):
    i = i+1
    y_pred = model.forward(x)
    print("---------")
    print(y_pred)
    loss = loss_function(y_pred,y)
    final_loss.append(loss)
    print("Epoch number:{}".format(i))
    print("----------WEIGHT---------")
    print(model.fc1.weight)
    print(model.fc2.weight)
    print("----------BIAS-----------")
    print(model.fc1.bias)
    print(model.fc2.bias)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print("----------LOSS----------")
    print(f"epoch = {i}, loss = {loss}")
```
